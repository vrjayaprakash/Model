```python
from transformers import pipeline
```

```python
summerizer = pipeline("summarization", device = 0)
text = """
The weather was unusually calm, and the sky looked clearer than ever before.
Technology continues to evolve, shaping the way we communicate and learn.
She opened the book and felt instantly drawn into its magical world.
The market was crowded, filled with the aroma of spices and fresh fruits.
Hard work pays off, but smart work pays off even faster.
I had an amazing day at work, everything went perfectly!
The new café in town has delicious food and a cozy atmosphere.

"""
summerizer(text, max_length = 30, min_length = 10)
```
